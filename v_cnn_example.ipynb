```python
from torchvision.models import resnet50, ResNet50_Weights
from v_cnn_extraction import create_feature_extractor, get_v_cnn, preprocess_image
```

```python
# Initialize ResNet50 pre-trained model
weights = ResNet50_Weights.DEFAULT
resnet = resnet50(weights=weights)
```

```python
# Create model from ResNet50 without the final fully connected layer
feature_extractor = create_feature_extractor(resnet)

# Preprocess the image
img = "data/ThumbnailTruthData/MTV_Thumbnails/XQmcH1yTbrA.jpg"
img_tensor = preprocess_image(img, weights)

# Get vector with visual features from the image
v_cnn = get_v_cnn(img_tensor, feature_extractor)
```
